```python
import time
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.firefox.options import Options
from selenium.webdriver.common.by import By
import json

# 1. Pegar Conteudo HTML a partir da URL
url = "https://www.clubefii.com.br/fundos-imobiliarios/20203/FOF"

option = Options()
option.headless = True
driver = webdriver.Firefox(options=option)

driver.get(url)
time.sleep(10)

html = driver.page_source
driver.quit()

soup = BeautifulSoup(html, "html.parser")

# Encontra a tabela com informações principais
tabela = soup.find("table", {"id": "tabela_profile"})

# 2. Pegar os títulos das colunas do <thead>
colunas = [th.text.strip() for th in tabela.find("thead").find_all("th")]

# 3. Extrair os dados do <tbody>
dados = []
for linha in tabela.find("tbody").find_all("tr"):
    cels = linha.find_all("td")
    linha_dados = [cel.get_text(strip=True) for cel in cels]
    dados.append(linha_dados)

# 4. Criar DataFrame com os dados extraídos
df = pd.DataFrame(dados, columns=colunas)

# 5. Exibir os dados dos administradores, por exemplo
print(colunas)
print(df[["Administrador", "Código Negociação", "Cotação", "Valor Último Rendimento", "Yield de Distribuição 1 Mês"]])

# Exportacao CSV
df.to_csv("fundos_imobiliarios.csv", index=False, encoding="utf-8-sig")

```
